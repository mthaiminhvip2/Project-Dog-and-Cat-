```python
from google.colab import drive
import os, shutil

# Mount Google Drive
drive.mount('/content/drive')

# Đường dẫn thư mục cha
parent_dir = "/content/drive/MyDrive/Project( Dog and Cat)/training_set"

# Duyệt qua các thư mục con
for subdir in os.listdir(parent_dir):
    subdir_path = os.path.join(parent_dir, subdir)
    if os.path.isdir(subdir_path):
        for file in os.listdir(subdir_path):
            src = os.path.join(subdir_path, file)
            dst = os.path.join(parent_dir, file)

            # Nếu trùng tên thì đổi tên file
            if os.path.exists(dst):
                name, ext = os.path.splitext(file)
                new_name = f"{name}_copy{ext}"
                dst = os.path.join(parent_dir, new_name)

            shutil.move(src, dst)

        os.rmdir(subdir_path)  # Xóa thư mục con sau khi trống
```
